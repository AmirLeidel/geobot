```python
import os
import shutil

# Define the folder path where the files are located
folder_path = data_path

# Loop through all the files in the folder
for file_name in os.listdir(folder_path):
    # Check if the file is a JSON file and is not a directory
    if file_name.endswith(".json") and os.path.isfile(os.path.join(folder_path, file_name)):
        # Get the parent folder path and the new folder path
        parent_folder_path = os.path.abspath(os.path.join(folder_path, os.pardir))
        new_folder_path = os.path.join(parent_folder_path, file_name)
        
        # Move the JSON file one folder up
        shutil.move(os.path.join(folder_path, file_name), new_folder_path)
        
        # Print a message to confirm the move
        print(f"Moved {file_name} to {new_folder_path}")
```

```python
with open(data_path+r"/sample.json") as f:
    data = json.load(f)["0"]
    print("Type:", type(data))
```

```python
unique_countries = set()
for d in data:
    if d['corrupted']:
      pass
    else:
      unique_countries.add(d['country'])
unique_countries = list(unique_countries)
num_countries = len(unique_countries)
print(num_countries)
```

```python
for d in data:
    if d['corrupted']:
      pass
    else:
      one_hot = [0] * num_countries
      country_index = unique_countries.index(d['country'])
      one_hot[country_index] = 1
      d['one_hot_country'] = one_hot
      
```
