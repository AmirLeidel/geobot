```python
import numpy as np
import requests

import json
```

```python
game_urls=[
    "https://www.geoguessr.com/game/tEUcR8jh3Fermr7b",
    "https://www.geoguessr.com/game/JIVcVjVJmeNM1ivG",
    "https://www.geoguessr.com/game/GJzecDoZyrYCZPFv"
    ]
ids = [url.split('/')[-1] for url in game_urls]
ids
```

```python
url = "https://geoguessr.com/api/v3/games/"
data = []
for id in ids:
  game_url = url+id
  response = requests.get(game_url)
  text = json.loads(response.text)['rounds']
  #result_ascii = []
  for round in text:
    round["panoId_ascii"] = bytearray.fromhex(round["panoId"]).decode()
    #esult_ascii.append(round)
    data.append(round)
  #data.append(result_ascii)
```
